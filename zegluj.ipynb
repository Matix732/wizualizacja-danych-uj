### Krok 1: Pobieranie danych z portalu Zegluj i zapis do CSV
Kod pliku zegluj_dane.py, przechodzi przez początkowe podstrony wyników, przetwarza zawartość HTML za pomocą biblioteki `BeautifulSoup` i wyodrębnia informacje takie jak nazwa jachtu, liczba osób, liczba kabin, rok produkcji, opis oraz cena.

Wyodrębnione dane zapisywane są w liście słowników, a potem konwertowane na DataFrame. Finalne dane gotowe do analizy przez poniższy plik są zapisywane w pliku `zegluj_dane.csv`.

### Krok 2: Wczytanie i korzystanie z zebranych danych z pliku
Aby zapobiec problemom z pobieraniem danych za każdym razem bezpośrednio ze strony - bazujemy na zebranych i zapisanych wcześniej danych w formacie CSV.

```python
import pandas as pd
csv_url = "https://raw.githubusercontent.com/Matix732/wizualizacja-danych-uj/main/zegluj_dane.csv"

# wczytanie danych csv 
df_z_pliku = pd.read_csv(csv_url)
```

### Krok 3: Analiza danych
Przystępujemy do analizy zebranych danych.

```python
import matplotlib.pyplot as plt

# Prosty Histogram pokazujący rozkład cen czarteru
df_z_pliku.hist(column='price', bins=10, color='lightgreen', edgecolor='black')

plt.xlabel('Cena czarteru')
plt.ylabel('Liczba jachtów z danego przedziału')
plt.title('Rozkład cen czarteru pobrany z pliku CSV')
plt.show()
```

### Średnia cena w zależności od rocznika produkcji
Wykres średniej ceny najmu za dobę w zależności od roku produkcji jachtu.

```python
# grupowanie po roku, liczenie średniej z cen
mean_by_year = df_z_pliku.groupby('year')['price'].mean().dropna()

plt.figure(figsize=(10, 5))
plt.bar(mean_by_year.index.astype(str), mean_by_year.values, color='cornflowerblue', edgecolor='black')
plt.xlabel('Rok produkcji')
plt.ylabel('Średnia cena czarteru (PLN)')
plt.title('Średnia cena za jacht w zależności od rocznika produkcji')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Średnia cena w zależności od ilości osób
Wykres ceny najmu jachtu w zależności od pojemności osób.

```python
# grupowanie po pojemności, liczenie średniej z cen
mean_by_people = df_z_pliku.groupby('people')['price'].mean().dropna()

plt.figure(figsize=(10, 5))
plt.plot(mean_by_people.index, mean_by_people.values, marker='o', color='crimson', linestyle='-', linewidth=2)
plt.xlabel('Ilość osób (pojemność)')
plt.ylabel('Średnia cena czarteru (PLN)')
plt.title('Średnia cena czarteru w zależności od ilości osób')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Średnia cena w zależności od ilości posiadanych kabin.
(odrzucamy dane gdzie nie ma informacji o ilości kabin tzn. = 0)

```python
# grupowanie po ilości kabin, liczenie średniej z cen
mean_by_cabins = df_z_pliku.groupby('cabins')['price'].mean().dropna()

plt.figure(figsize=(10, 5))
plt.bar(mean_by_cabins.index.astype(str), mean_by_cabins.values, color='mediumseagreen', edgecolor='black')
plt.xlabel('Ilość kabin')
plt.ylabel('Średnia cena czarteru (PLN)')
plt.title('Średnia cena czarteru w zależności od ilości kabin')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Wykres ilości kabin dla poszczególnych pojemności (liczby osób)


```python
plt.figure(figsize=(10, 5))
plt.scatter(df_z_pliku['people'], df_z_pliku['cabins'], color='purple', alpha=0.5, s=100)
plt.xlabel('Ilość osób (pojemność całkowita)')
plt.ylabel('Ilość kabin')
plt.title('Zależność ilości kabin od pojemności jachtu')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
